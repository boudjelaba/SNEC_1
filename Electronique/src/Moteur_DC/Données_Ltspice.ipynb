# Importation des données .txt dans Python

## Méthode 1 :

```python
import numpy  as np
import matplotlib.pyplot as plt
%matplotlib inline

data = np.loadtxt('Donnees.txt', skiprows=1)

x = data[:,0]
y = data[:,1]

plt.figure(figsize=(8,5))
plt.plot(x, y,'r-')
plt.grid()
plt.show()
```

## Méthode 2 :

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('Donnees.txt', skiprows=1, sep='\t')
df.to_csv('Moteur_DC1.csv',index=None)
df1 = pd.read_csv('Moteur_DC1.csv', names=['t','v'])
x1 = df1.t
y1 = df1.v

plt.figure(figsize=(8,5))
plt.plot(x1, y1,'b-')
plt.grid()
plt.show()
```
